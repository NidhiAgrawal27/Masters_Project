```python
columns=["transaction_num","id_input_addresses_x","id_output_addresses_y","input_amounts_x","output_amounts_y"]
data=data[columns]
data.info()
```

To implement the change address heuristics, the steps are as follows:
- Firstly, consider dictionaries of all transactions in which an address is an input or output.
- Second, consider dictionaries of all transaction input and output amounts.
- While iterating through the input and output amounts- check in which transactions are common for such two i/o addresses.
- Move to the dictionary where the input and output amounts are listed. Here, there are subconditions:

- a. If the output amount is less than the sum of the inputs, it is most likely the change address. This argument is because many inputs may combine to pay a higher amount to another output user. The rest is the change.
- b. If there is a single input and output user and the entire amount is being transferred, most likely there is a selling that has happened? So should not be considered a heuristic.
- c. For a single input, the one with the lower change should be considered the change address or higher? Both scenarios fit. Very vague approach.

```python
#implement the heuristic
list=[]
for index,(inp_add, out_add, in_amt, out_amt) in enumerate(zip(data["id_input_addresses_x"],data["id_output_addresses_y"],data["input_amounts_x"],data["output_amounts_y"])):
    if ((inp_add, out_add) or (out_add,inp_add)) in list:
        continue
    else:
        list.append((inp_add, out_add))
        transaction_list=set(input_add_to_tran[inp_add]).intersection(output_add_to_tran[out_add])
        for transaction in transaction_list:
            #condition 1- (for a single input address), multiple output add---- output amount has to be less than all the input amounts.
            #if (len(tranId_to_inpadd[transaction])==1 and len(tranId_to_outadd[transaction])>1):
            if (len(tranId_to_outadd[transaction])>1):
                #if output amount is the smallest of all amounts
                if (out_amt<np.sum(tranId_to_inpamounts[transaction]) and out_amt<(sorted(tranId_to_inpamounts[transaction])[0])):   #smaller than the least input amount
                    data.loc[((data["id_input_addresses_x"]==inp_add) & (data["id_output_addresses_y"]==out_add)) | ((data["id_input_addresses_x"]==out_add) & (data["id_output_addresses_y"]==inp_add)),"h1"]=1
                    break
                else:
                    data.loc[((data["id_input_addresses_x"]==inp_add) & (data["id_output_addresses_y"]==out_add)) | ((data["id_input_addresses_x"]==out_add) & (data["id_output_addresses_y"]==inp_add)),"h1"]=0
            else:
                data.loc[((data["id_input_addresses_x"]==inp_add) & (data["id_output_addresses_y"]==out_add)) | ((data["id_input_addresses_x"]==out_add) & (data["id_output_addresses_y"]==inp_add)),"h1"]=0
```
