### Heuristic 0- All input addressses belong to the same user.

```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("segregated_iota.csv")
data.head()
```

```python
columns=["transaction_id","input_addresses_x","output_addresses_y","id_input_addresses_x","id_output_addresses_y"]
data=data[columns]
data.head()
```

```python
df_tx_id = pd.DataFrame()
df_tx_id['transaction_id'] = pd.unique(data[["transaction_id"]].values.ravel())
df_tx_id['transaction_num'] = df_tx_id.index
df_tx_id.shape
```

```python
data = pd.merge(data,df_tx_id, on='transaction_id', how='left')
data.head()
```

```python
data[data["transaction_num"]==2435]
```

```python
df=data[["transaction_num","id_input_addresses_x","id_output_addresses_y"]]
df
```

```python
#map step
# sample=df.iloc[:10,:]
# print(sample)

add_map_tx={}
for address_id in np.unique(df["id_input_addresses_x"]):
    new_df=df[df["id_input_addresses_x"]==address_id]
    add_map_tx[address_id]=set(new_df["transaction_num"])
print(len(add_map_tx[0]))
print(add_map_tx[37])
print(add_map_tx[38])
```

```python
#reduce step- unweighted

for index,(input,output) in enumerate(zip(df["id_input_addresses_x"],df["id_output_addresses_y"])):
    if output not in add_map_tx.keys():
        df.loc[index,"h0"]=0
        df.loc[index,"wh0"]=0
    else:
        if (len(add_map_tx[input].intersection(add_map_tx[output]))!=0):
            df.loc[index,"h0"]=1
            df.loc[index,"wh0"]=len(add_map_tx[input].intersection(add_map_tx[output]))
        else:
            df.loc[index,"h0"]=0
            df.loc[index,"wh0"]=0
```
